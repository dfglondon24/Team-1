```python
import pandas as pd
import os

df_learner = []
for i in range(1, 3):
    file_path = fr'C:\Users\thoma\OneDrive\Bureau\hacka\data1\Cycle_{i}.xls'
    if os.path.exists(file_path):
        print("number", i)
        df = pd.read_excel(file_path, sheet_name=1)
        df_filtered = df.query('IP == "AfriKids"')
        df_filtered = df_filtered[['IP', 'REGION', 'DISTRICT', 'COMMUNITY/CLASS', 'TYPE OF OOSC', 'SEX',
                                   'AGE', 'DROP OUT (YES/NO)', 'WHICH QUARTER CHILD DROPPED',
                                   'GRADUATED (YES/NO)', 'TRANSITIONED (YES/NO)', 'SCHOOL TRANSITIONED',
                                   'CLASS PLACED IN', 'LANGUAGE']]
        df_learner.append(df_filtered)
        print(df_filtered)


for i in range(3, 4):
    file_path = fr'C:\Users\thoma\OneDrive\Bureau\hacka\data1\Cycle_{i}.xlsx'
    if os.path.exists(file_path):
        print("number", i)
        df = pd.read_excel(file_path, sheet_name=1)
        df_filtered = df.query('IP == "Afrikids"')
        df_filtered = df_filtered[['IP', 'REGION', 'DISTRICT', 'COMMUNITY/CLASS', 'TYPE OF OOSC', 'SEX',
                                   'AGE', 'DROP OUT (YES/NO)', 'WHICH QUARTER CHILD DROPPED',
                                   'GRADUATED (YES/NO)', 'TRANSITIONED (YES/NO)', 'SCHOOL TRANSITIONED',
                                   'CLASS PLACED IN', 'LANGUAGE']]
        df_learner.append(df_filtered)
        print(df_filtered)

for i in range(4, 6):
    file_path = fr'C:\Users\thoma\OneDrive\Bureau\hacka\data1\Cycle_{i}.xls'
    if os.path.exists(file_path):
        print("number", i)
        df = pd.read_excel(file_path, sheet_name=0)
        df_filtered = df.query('IP == "Afrikids"')
        df_filtered = df_filtered[['IP', 'REGION', 'DISTRICT', 'COMMUNITY/CLASS', 'TYPE OF OOSC', 'SEX',
                                   'AGE', 'DROP OUT (YES/NO)', 'WHICH QUARTER CHILD DROPPED',
                                   'GRADUATED (YES/NO)', 'TRANSITIONED (YES/NO)', 'SCHOOL TRANSITIONED',
                                   'CLASS PLACED IN', 'LANGUAGE']]
        df_learner.append(df_filtered)
        print(df_filtered)

df_combined = pd.concat(df_learner, ignore_index=True)
df_combined
```
